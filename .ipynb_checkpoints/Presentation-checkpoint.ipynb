## Data Preprocessing

```python
train = FashionMNIST('data', download= True, train=True, transform=transforms.ToTensor())
test = FashionMNIST('data', download = True, train = False, transform=transforms.ToTensor())
```

```python
batch_size = 100

train_loader = torch.utils.data.DataLoader(dataset=train, batch_size=batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test, batch_size=batch_size, shuffle=True)
```

```python
class FashionBaselineModel():
    def __init__(self, labels, data):
        self.unique_labels = torch.unique(labels)
        self.average_images = {}
        
        for label in self.unique_labels:
            self.average_images[label.item()] = data[labels==label].sum(dim=0)/ data[labels==label].shape[0]
        print('Trained Baseline Model.')
            
    def inference_on_image(self, image):
        performance = {}
        for label in self.unique_labels:
            performance[label.item()] = torch.sum(torch.abs(self.average_images[label.item()] - image))
        return min(performance, key=performance.get)

    def score_on_dataset(self, labels, data):
        n_correct = 0 
        for i in range(len(labels)):
            target_label = labels[i]
            is_correct = self.inference_on_image(data[i]) == target_label
            
            if is_correct:
                n_correct += 1
        return n_correct / len(labels)
                    
                
                
```

```python
baseline = FashionBaselineModel(train.targets, train.data)
performance = baseline.score_on_dataset(test.targets, test.data)
print('Baseline Performance on Test Set', performance)
```

# Training a Neural Network

```python
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(784, 128)  # Input size (28x28 images) = 784
        self.fc2 = nn.Linear(128, 10)   # Output size (digits 0-9) = 10

    def forward(self, x):
        x = x.view(x.shape[0], -1) # x.shape[0] is the number of images in a batch
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
    
simple_network = SimpleNN()
```

```python
learning_rate = 0.01

def train_network(n_epochs, loader, network):
    n_total_steps = len(loader)
    optimizer = optim.SGD(network.parameters(), lr=learning_rate)
    loss_criterion = nn.CrossEntropyLoss()
    for epoch in range(n_epochs):
        for i, (images, labels) in enumerate(loader):
            optimizer.zero_grad()
            output = network(images)
            loss = loss_criterion(output, labels)
            loss.backward()
            optimizer.step()
            if (i+1) % 100 == 0:
                print (f'Epoch [{epoch+1}/{n_epochs}], Step [{i+1}/{n_total_steps}], Loss: {loss.item():.4f}')
    print('Training Finished')
    return network
```

```python
def test_network(loader, network):
    with torch.no_grad():
        n_correct = 0
        n_samples = len(loader.dataset)
        for images, labels in loader:

            outputs = network(images)

            # max returns (output_value ,index)
            _, predicted = torch.max(outputs, 1)
            n_correct += (predicted == labels).sum().item()

        acc = n_correct / n_samples
        print(f'Accuracy of the network on the {n_samples} test images: {100*acc} %')
```

```python
simple_network = train_network(3, train_loader, simple_network)
test_network(test_loader, simple_network)
```

```python
class FashionCNN(nn.Module):
    def __init__(self):
        super(FashionCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(64 * 7 * 7, 128)
        self.fc2 = nn.Linear(128, 10)
        self.dropout = nn.Dropout(0.25)  # 25% dropout rate

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.dropout(x)  # Apply dropout after the first convolutional layer
        x = self.pool(F.relu(self.conv2(x)))
        x = self.dropout(x)  # Apply dropout after the second convolutional layer
        x = x.view(-1, 64 * 7 * 7)
        x = F.relu(self.fc1(x))
        x = self.dropout(x)  # Apply dropout after the first fully connected layer
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
cnn = FashionCNN()
```

```python
cnn = train_network(10, train_loader, cnn)
test_network(test_loader, cnn)
```

```python
class ImprovedFashionCNN(nn.Module):
    def __init__(self):
        super(ImprovedFashionCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(128 * 3 * 3, 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 10)
        self.dropout = nn.Dropout(0.25)
        
    def forward(self, x):
        x = self.pool(self.bn1(F.relu(self.conv1(x))))
        x = self.pool(self.bn2(F.relu(self.conv2(x))))
        x = self.pool(self.bn3(F.relu(self.conv3(x))))
        x = x.view(-1, 128 * 3 * 3)
        x = self.dropout(F.relu(self.fc1(x)))
        x = self.dropout(F.relu(self.fc2(x)))
        x = self.fc3(x)
        return F.log_softmax(x, dim=1)
improved_cnn = ImprovedFashionCNN()
```

```python
improved_cnn = train_network(10, train_loader, improved_cnn)
test_network(test_loader, improved_cnn)
```
